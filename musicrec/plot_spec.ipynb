```python
import librosa, matplotlib.pyplot as plt, numpy as np, librosa.display
```

```python
musicfile = '/home/carsten/workspaces/isws/audio/testfiles/GTZAN/genres/classical/classical.00000.wav'
```

```python
y, sr = librosa.load(musicfile)
```

```python
plt.figure(figsize=(14, 5))
plt.title('Waveplot - classical.00000.wav')
librosa.display.waveplot(y, sr=sr)
```

```python
Y = librosa.stft(y)
Ydb = librosa.amplitude_to_db(abs(Y))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Ydb, sr=sr, x_axis='time', y_axis='hz')
plt.title('Spectrogram - classical.00000.wav')
plt.colorbar()
```

```python
y, sr = librosa.load(musicfile)
plt.figure(figsize=(14, 5))

S = librosa.feature.melspectrogram(y=y, sr=sr)

librosa.display.specshow(librosa.power_to_db(S,ref=np.max),
                         y_axis='mel', fmax=8000,
                        x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram - classical.00000.wav')
plt.tight_layout()
```
